# Levenshtein Distance Algorithm

The Levenshtein distance measures the minimum number of single-character edits (insertions, deletions, or substitutions) needed to change one string into another.

**Algorithm**: Dynamic Programming approach with $O(m \cdot n)$ time complexity.

```python
def levenshtein_distance(s1, s2):
	"""
	Calculate the Levenshtein distance between two strings.
	
	Args:
		s1 (str): First string
		s2 (str): Second string
	
	Returns:
			int: The minimum number of edits needed
	"""
	m, n = len(s1), len(s2)
	
	# Create a DP table
	dp = [[0] * (n + 1) for _ in range(m + 1)]
	
	# Initialize base cases
	for i in range(m + 1):
		dp[i][0] = i  # Cost of deleting all characters from s1
	
	for j in range(n + 1):
		dp[0][j] = j  # Cost of inserting all characters to make s2
	
	# Fill the DP table
	for i in range(1, m + 1):
		for j in range(1, n + 1):
			if s1[i - 1] == s2[j - 1]:
				# Characters match, no additional cost
				dp[i][j] = dp[i - 1][j - 1]
			else:
				# Take minimum of three operations
				dp[i][j] = 1 + min(
					dp[i - 1][j],      # Deletion
					dp[i][j - 1],      # Insertion
					dp[i - 1][j - 1]   # Substitution
				)
	
	return dp, dp[m][n]
```

### **Test Examples**

```python
# Example 1: Similar words
word1 = "kitten"
word2 = "sitting"
_, distance = levenshtein_distance(word1, word2)
print(f"Levenshtein distance between '{word1}' and '{word2}': {distance}")

# Example 2: Identical strings
word3 = "hello"
word4 = "hello"
_, distance2 = levenshtein_distance(word3, word4)
print(f"Levenshtein distance between '{word3}' and '{word4}': {distance2}")

# Example 3: Completely different
word5 = "abc"
word6 = "xyz"
_, distance3 = levenshtein_distance(word5, word6)
print(f"Levenshtein distance between '{word5}' and '{word6}': {distance3}")

# Example 4: One empty string
word7 = "test"
word8 = ""
_, distance4 = levenshtein_distance(word7, word8)
print(f"Levenshtein distance between '{word7}' and '{word8}': {distance4}")
```

### **Visualization: DP Table**

Let's visualize the DP table to understand how the algorithm works.

```python
s1 = "kitten"
s2 = "sitting"
m, n = len(s1), len(s2)

dp, distance = levenshtein_distance(s1, s2)

# Print the table
print(f"      {' '.join(f'{c:3}' for c in [''] + list(s2))}")
for i in range(m + 1):
	row_label = '' if i == 0 else s1[i - 1]
	print(f"{row_label:3} {' '.join(f'{dp[i][j]:3}' for j in range(n + 1))}")
print(f"\nFinal distance: {distance}")
```
