#### **YouTube IDs**

Have you ever wondered how **YouTube** generates those unique video IDs without even running out of combinations? Let's first understand the basics.

YouTube uses a Base 64 character set, which includes:
- **Uppercase letters:** A-Z (26 characters)
- **Lowercase letters:** a-z (26 characters)
- **Digits:** 0-9 (10 characters)
- **Special characters:** '-' and '_' (2 characters)

This gives us a total of 64 unique characters. Each YouTube video ID is typically 11 characters long.

To calculate the total number of unique combinations for an 11-character ID using Base 64, we use the formula:
$$\text{Total Combinations} = 64^{11} = 73,786,976,294,838,206,464$$

This means YouTube can generate over 73 quintillion unique video IDs, which is more than enough to accommodate the billions of videos uploaded to the platform without running out of combinations!

And if it ever did approach that limit, YouTube could simply increase the length of the video IDs to create even more combinations. For example, a 12-character ID would yield:
$$\text{Total Combinations} = 64^{12} = 4,722,366,482,869,645,213,696$$
and so on.

```python
import random
import string
```

```python
class YouTubeID:
	def __init__(self, length=11):
		# Base 64 characters: 0-9, A-Z, a-z, hyphen, underscore
		self.base64_chars = string.digits + string.ascii_uppercase + string.ascii_lowercase + '-' + '_'
		self.used_ids = set()  # Track used IDs
		self.length = length

	def generate_id(self):
		"""Generate a random Base 64 ID"""
		return ''.join(random.choice(self.base64_chars) for _ in range(self.length))

	def upload_video(self, title):
		"""Simulate uploading a video and getting a unique ID"""
		max_attempts = 100
		
		for attempt in range(max_attempts):
				video_id = self.generate_id()
				
				# Check if ID is already taken
				if video_id not in self.used_ids:
					self.used_ids.add(video_id)
					print(f"✓ Video '{title}' uploaded!")
					print(f"  ID: {video_id}")
					print(f"  URL: https://youtube.com/watch?v={video_id}\n")
					return video_id
		
		print("Failed to generate unique ID (very unlikely!)")
		return None

	def calculate_total_combinations(self, length=None):
		"""Calculate total possible IDs"""
		if length is None:
			length = self.length
		
		total = 64 ** length
		return total
```

```python
# Example Usage
youtube = YouTubeID(length=8)

# Upload some videos
youtube.upload_video("My First Video")
youtube.upload_video("Generative AI Explained")
youtube.upload_video("Tutorial: Python Basics")

# Show statistics
print(f"Total videos uploaded: {len(youtube.used_ids)}")

# Calculate total possible combinations
total = youtube.calculate_total_combinations()
print(f"Total possible {youtube.length}-character IDs: {total:,}")
```
