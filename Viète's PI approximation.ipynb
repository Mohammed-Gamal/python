Viète's formula is indeed a beautiful piece of mathematical history! Discovered in 1593, it was the first exact formula for $\pi$ expressed as an infinite product, and it is elegant in how it relies purely on the number 2 and nested square roots.

To translate this into code, we first need to look at the pattern as a recursive sequence. We can define the denominator terms, let's call them $a_n$, like this:

$$a_1 = \sqrt{2}$$
$$a_{n} = \sqrt{2 + a_{n-1}}$$

Then, the formula for $\pi$ can be expressed as:
$$
\begin{align*}
	\pi &\approx 2 \cdot \frac{2}{\sqrt{2}} \cdot \frac{2}{\sqrt{2 + \sqrt{2}}} \cdot \frac{2}{\sqrt{2 + \sqrt{2 + \sqrt{2}}}} \cdots  \\\\
	&\approx 2 \cdot \frac{2}{a_1} \cdot \frac{2}{a_2} \cdot \frac{2}{a_3} \cdots \\ \\
	&\approx 2 \cdot \prod_{n=1}^{\infty} \frac{2}{a_n}
\end{align*}
$$

---
### **Standard Python Approach**
- Suffers from floating-point precision issues as the number of iterations increases, leading to a loss of accuracy in the approximation of $\pi$.

```python
import math

def viete_pi_standard(iterations=20):
	pi_approx = 2.0
	a_n = math.sqrt(2.0)
	
	for _ in range(iterations):
		pi_approx *= (2.0 / a_n)
		a_n = math.sqrt(2.0 + a_n)  # Calculate the next denominator
	
	return pi_approx

# Test the function
approx = viete_pi_standard(25)
print(f"Approximation:  {approx}")
print(f"Actual math.pi: {math.pi}")
print(f"\nError: {math.pi - approx}")
```

---
### **High-Precision Approach with mpmath**

- By using the `mpmath` library, we can set a high number of decimal places for our calculations, allowing us to compute a much more accurate approximation of $\pi$.

```python
import mpmath

# Set the number of decimal places (dps)
mpmath.mp.dps = 60 

def viete_pi_mpmath(iterations=100):
	pi_approx = mpmath.mpf(2.0)
	a_n = mpmath.sqrt(2.0)
	
	for _ in range(iterations):
		pi_approx *= (2.0 / a_n)
		a_n = mpmath.sqrt(2.0 + a_n)
	
	return pi_approx

print(f"Approx: {viete_pi_mpmath(1500)}")
print(f"Actual: {mpmath.mp.pi}")
print(f"\nError: {mpmath.absmax(mpmath.mp.pi - viete_pi_mpmath(1500))}")
```
